목표 : 시장지표 만들기
========
- PER = 주가 / (당기순이익 / 주식수) = 주가 / 주당순이익 = 시가총액 / 당기순이익 = 주가 * 주식수 / 당기순이익
- PBR = 주가 / (순자산 / 주식수) = 주가 / 주당순자산 = 시가총액 / 순자산(자본총계) = 주가 * 주식수 / 순자산
- PCR = 주가 / (영업현금흐름 / 주식수) = 시가총액 / 영업현금흐름 = 주가 * 주식수 / 영업현금
- PSR = 주가 / (매출액 / 주식수) = 시가총액 / 매출액 = 주가 * 주식수 / 매출액
