```python
import pandas as pd
```

```python
kskqkn = pd.read_csv("../1.데이터셋/Feature/상장피처clean2.csv")
kskqkn.columns
```

```python
# 각 행의 0의 개수를 계산하여 'zero_count' 열에 저장
kskqkn["zero_count"] = kskqkn.apply(lambda row: row.value_counts().get(0, 0), axis=1)
```

```python
kskqkn['zero_count'].quantile(0.8)
```

```python
kskqkn['zero_count'].describe()
```

```python
kskqkn = kskqkn[kskqkn["zero_count"] <= 6]
```

```python
kskqkn = kskqkn.drop(["zero_count"], axis=1)
```

```python
# get_dummies() 함수를 사용하여 '기업수명주기' 열을 더미 변수로 변환
df_dummy = pd.get_dummies(kskqkn["기업수명주기"], prefix="기업수명주기")

# 데이터프레임에 더미 변수 추가
kskqkn = pd.concat([kskqkn, df_dummy], axis=1)

# # '기업수명주기' 열을 정수형으로 변환 (0 또는 1)
# kskqkn["기업수명주기"] = kskqkn["기업수명주기"].astype(int)

# '기업수명주기' 열을 더미 변수로 변환하기 전 열 제거
kskqkn.drop(columns=["기업수명주기"], axis=1, inplace=True, errors="ignore")
```

```python
# '기업수명주기' 열을 정수형으로 변환 (0 또는 1)
kskqkn[["기업수명주기_도입기", "기업수명주기_성숙기", "기업수명주기_성장기", "기업수명주기_쇠퇴기"]] = kskqkn[
    ["기업수명주기_도입기", "기업수명주기_성숙기", "기업수명주기_성장기", "기업수명주기_쇠퇴기"]
].astype(int)
```
