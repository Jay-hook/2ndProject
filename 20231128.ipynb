```python
import pandas as pd
import numpy as np
import platform
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
if platform.system() == "Darwin":           # MAC인 경우
    plt.rc('font', family ="AppleGothic")
else:
    plt.rc('font',family='Malgun Gothic')

#멧플로립에서 음수를 표시
plt.rc('axes',unicode_minus=False)

#워닝 메시지
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./새 폴더/외감_재무비율.csv', encoding='utf-8',  index_col=0)
```

```python
외감_재무비율_성장성 = df.iloc[:,3:17]
외감_재무비율_성장성.isnull().sum()

#외감_재무비율_성장성.info()
#외감_재무비율_성장성.describe()
외감_재무비율_성장성.drop('경상이익증가율 (2007년 이전 발생)' ,axis=1, inplace=True)
외감_재무비율_성장성.eq(0).sum(axis=0)
```

```python
import scipy.stats as stats
from scipy.stats import shapiro

for i in 외감_재무비율_성장성.columns:
    statistic, pvalue = stats.shapiro(외감_재무비율_성장성[i])
    print(f'{i}: {statistic:.4f}, {pvalue:.4f}')


```

```python
plt.figure(figsize=(10, 8))
mask = np.triu(np.ones_like(외감_재무비율_성장성.corr(), dtype=bool))



sns.heatmap(외감_재무비율_성장성.corr(), annot=True, cmap='coolwarm', linewidths=.5, fmt='.2f',mask=mask)
```

```python
외감_재무비율_수익성 = df.iloc[:,18:67]
외감_재무비율_수익성.isnull().sum()
외감_재무비율_수익성.eq(0).sum(axis=0)
# 데이터수집기간이 12년도부터라서 아래 계정은 삭제함 
외감_재무비율_수익성.drop('매출액경상이익률 (2007년 이전 발생)' ,axis=1, inplace=True)
외감_재무비율_수익성.drop('총자본경상이익률 (2007년 이전 발생)' ,axis=1, inplace=True)
외감_재무비율_수익성.drop('자기자본경상이익률 (2007년 이전 발생)' ,axis=1, inplace=True)
외감_재무비율_수익성.drop('자본금경상이익률 (2007년 이전 발생)' ,axis=1, inplace=True)
외감_재무비율_수익성.drop('기업경상이익률 (2007년 이전 발생)' ,axis=1, inplace=True)
외감_재무비율_수익성.drop('투자수익률 (2007년 이전 발생)' ,axis=1, inplace=True)
외감_재무비율_수익성.drop('1주당경상이익 (2007년 이전 발생)(원)',axis=1, inplace=True)
```
