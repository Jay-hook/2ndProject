```python
import pandas as pd
```

```python
market_data = pd.read_csv('../1.데이터셋/시장지표/상장_시장지표.csv', index_col=0)
market_data
```

```python
# '회계년도' 열을 datetime 형식으로 변환
market_data["회계년도"] = pd.to_datetime(market_data["회계년도"], format="%Y/%m")
# 년도만 추출
market_data["회계년도"] = market_data["회계년도"].dt.year
```
