```python
import pandas as pd
import os

pd.options.display.float_format = "{:.5f}".format

unlisted_directory = "../1.데이터셋/재무비율/외감재무비율"
os.listdir(unlisted_directory)
```

```python
result = pd.DataFrame()

for filename in os.listdir(unlisted_directory):
    df = pd.read_csv(f"../1.데이터셋/재무비율/외감재무비율/{filename}", encoding="cp949")
    result = pd.concat([result, df])
print(result)
```

```python
result["회계년도"] = pd.to_datetime(result["회계년도"])
result = result[result["회계년도"].dt.month == 12].reset_index(drop=True)
result["회계년도"] = result["회계년도"].dt.year
```
