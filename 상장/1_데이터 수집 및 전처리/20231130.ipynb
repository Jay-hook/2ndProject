```python
import statsmodels.api as sm
X = df.drop(['부실판단', '기업수명주기'], axis=1)
y = df[['부실판단']]

print(X.dtypes)
print(y.dtypes)
```

```python
# OLS 정보 보기 
feature_add = sm.add_constant(X, has_constant='add')
# sm OLS 적합
model = sm.OLS(y , feature_add)
fitted_model = model.fit()

# summary 함수통해 결과출력
fitted_model.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

df1 = pd.concat([X, y], axis=1)

# Assuming df1 is your DataFrame
numeric_cols = df1.select_dtypes(include='number').columns.tolist()
numeric_cols.remove('부실판단')  # Remove the target variable from the list

# Calculate VIF for each feature
VIF_data = pd.DataFrame()
VIF_data['feature'] = numeric_cols
VIF_data['VIF'] = [variance_inflation_factor(df1[numeric_cols].values, i) for i in range(len(numeric_cols))]

# Display the VIF values
print(VIF_data.sort_values(by='VIF', ascending=False))
```

```python
# 종속변수와 인과관계 확인 
np.abs(df1.corr()['부실판단']).sort_values(ascending=False)
```

```python


# Assuming df1 is your DataFrame
outliers_percentage = []

for k, v in df1.items():
    q1 = v.quantile(0.25)
    q3 = v.quantile(0.75)
    irq = q3 - q1
    v_col = v[(v <= q1 - 1.5 * irq) | (v >= q3 + 1.5 * irq)]
    perc = np.shape(v_col)[0] * 100.0 / np.shape(df1)[0]
    outliers_percentage.append((k, perc))

# Sort the list of tuples by percentage in descending order
outliers_percentage.sort(key=lambda x: x[1], reverse=True)

# Display the sorted results
for col, perc in outliers_percentage:
    print(f"Column {col} outliers = {perc:.2f}%")
```

```python
import matplotlib.pyplot as plt
from scipy import stats  # stats 모듈을 import 추가

def qq_plot(df):
    # 모든 칼럼 순회
    for col in df.columns:  # 변수명을 df로 수정 (df1 -> df)
        # Q-Q plot 그리기
        stats.probplot(df[col].dropna(), dist="norm", plot=plt)
        plt.title(f"Q-Q plot for {col}")
        plt.show()


qq_plot(df1)
```
