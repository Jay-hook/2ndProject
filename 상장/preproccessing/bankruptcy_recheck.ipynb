```python
import pandas as pd

pd.options.display.float_format = "{:.2f}".format
```

```python
kskqkn = pd.read_csv("../1.데이터셋/부실/상장_영업이익/상장_영업이익.csv", index_col=0)
kskqkn = kskqkn[
    [
        "회사명",
        "거래소코드",
        "회계년도",
        "이자보상배율(이자비용)",
        "[U01250010]   이자비용(천원)",
        "[U01230000]영업이익(손실)(천원)",
    ]
]

audit = pd.read_csv("../1.데이터셋/부실/상장_영업이익/외감_영업이익.csv", index_col=0)
audit = audit[
    [
        "회사명",
        "거래소코드",
        "회계년도",
        "이자보상배율(이자비용)",
        "[U01250010]   이자비용(천원)",
        "[U01230000]영업이익(손실)(천원)",
    ]
]
audit
```

```python
kskqkn["회계년도"] = pd.to_datetime(kskqkn["회계년도"])
kskqkn = kskqkn[kskqkn["회계년도"].dt.month == 12].reset_index(drop=True)
kskqkn["회계년도"] = kskqkn["회계년도"].dt.year

audit["회계년도"] = pd.to_datetime(audit["회계년도"])
audit = audit[audit["회계년도"].dt.month == 12].reset_index(drop=True)
audit["회계년도"] = audit["회계년도"].dt.year
```

```python
kskqkn.loc[
    (kskqkn["이자보상배율(이자비용)"] == 999999999) & (kskqkn["[U01230000]영업이익(손실)(천원)"] < 0),
    "이자보상배율(이자비용)",
] = 0

audit.loc[
    (audit["이자보상배율(이자비용)"] == 999999999) & (audit["[U01230000]영업이익(손실)(천원)"] < 0),
    "이자보상배율(이자비용)",
] = 0
```

```python
df = pd.read_csv("../1.데이터셋/부실/부실_상장.csv", encoding="cp949")
df.head(3)

df2 = pd.read_csv("../1.데이터셋/부실/외감_부실.csv", encoding='cp949', index_col=0)
df2.head(3)
```

```python
df["회계년도"] = pd.to_datetime(df["회계년도"], format="%b-%y")
df = df[df["회계년도"].dt.month == 12].reset_index(drop=True)
df["회계년도"] = df["회계년도"].dt.year

df2["회계년도"] = pd.to_datetime(df2["회계년도"], format="%b-%y")
df2 = df2[df2["회계년도"].dt.month == 12].reset_index(drop=True)
df2["회계년도"] = df2["회계년도"].dt.year
```

```python
df2
```

```python
listed_merged_df = pd.merge(
    df, kskqkn[["거래소코드", "회계년도", "이자보상배율(이자비용)"]], on=["거래소코드", "회계년도"], how="left"
)


unlisted_merged_df = pd.merge(
    df2, audit[["거래소코드", "회계년도", "이자보상배율(이자비용)"]], on=["거래소코드", "회계년도"], how="left"
)
```

```python
listed_merged_df.columns
unlisted_merged_df.columns
```

```python
listed_df = listed_merged_df[
    [
        "회사명",
        "거래소코드",
        "회계년도",
        "감사의견코드",
        "[U01410000]영업활동으로 인한 현금흐름(*)(천원)",
        "이자보상배율(이자비용)_y",
    ]
]

unlisted_df = unlisted_merged_df[
    [
        "회사명",
        "거래소코드",
        "회계년도",
        "감사의견코드",
        "[U01410000]영업활동으로 인한 현금흐름(*)(천원)",
        "이자보상배율(이자비용)_y",
    ]
]
```

```python
listed_df.rename(columns={"이자보상배율(이자비용)_y": "이자보상배율(이자비용)"}, inplace=True)

unlisted_df.rename(columns={"이자보상배율(이자비용)_y": "이자보상배율(이자비용)"}, inplace=True)
unlisted_df
```

```python
listed_df.to_csv("../1.데이터셋/부실/new_상장_부실.csv")
unlisted_df.to_csv("../1.데이터셋/부실/new_외감_부실.csv")
```
