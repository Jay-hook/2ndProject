```python
audit_fin = pd.read_csv("../1.데이터셋/재무비율/외감_재무비율_final.csv")
kskqkn_fin = pd.read_csv("../1.데이터셋/재무비율/상장_재무비율_j_rev01.csv")
audit_fin_plus = pd.read_csv("../1.데이터셋/재무비율/누락데이터.csv", encoding="cp949")
audit_lc = pd.read_csv("../1.데이터셋/Life_Cycle_Label/비상장부실+수명주기.csv", index_col=0)
kskqkn_lc = pd.read_csv("../1.데이터셋/Life_Cycle_Label/상장부실+수명주기.csv", index_col=0)
```
