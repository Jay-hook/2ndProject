```python
import pandas as pd
import numpy as np
# pd.set_option('display.max_rows', None)
pd.reset_option('display.max_rows')
```

```python
df1 = pd.read_csv('./datasets/상장_부실라벨링작업중.csv', encoding = 'cp949')
df2 = pd.read_csv('./datasets/비상장_부실라벨링작업중.csv', encoding='utf-8')
```

```python
# 상장 부실판단
df1['부실판단'] = None
cond1 = (df1['감사부실'] == 1) | (df1['이보배부실'] == 1) | (df1['현금흐름부실'] == 1)
cond2 = (~cond1) & (df1['감사부실'] == 0) | (df1['이보배부실'] == 0) | (df1['현금흐름부실'] == 0)

df1.loc[cond1, '부실판단'] = 1
df1.loc[cond2, '부실판단'] = 0

# 비상장 부실판단
df2['부실판단'] = None
cond1 = (df2['감사부실'] == 1) | (df2['이보배부실'] == 1) | (df2['현금흐름부실'] == 1)
cond2 = (~cond1) & (df2['감사부실'] == 0) | (df2['이보배부실'] == 0) | (df2['현금흐름부실'] == 0)

df2.loc[cond1, '부실판단'] = 1
df2.loc[cond2, '부실판단'] = 0
```

```python
## 상장기업
# 부실기업 데이터만 출력
bank1 = df1[df1['부실판단'] == 1]
bank1.drop_duplicates(subset='거래소코드', keep='first', inplace=True)

# 정상기업 데이터만 남기기
bank1_code = bank1['거래소코드'].unique() # 부실로 선정된 기업의 거래소코드
non_bank1 = df1[~df1['거래소코드'].isin(bank1_code)] # 부실선정 기업 제거
non_bank1.drop_duplicates(subset='거래소코드', keep='last', inplace=True)

# 부실기업과 정상기업 데이터프레임 합치기
final1 = pd.concat([bank1, non_bank1])
final1.reset_index(drop=True, inplace=True)


### 비상장 기업
# 부실기업 데이터만 출력
bank2 = df2[df2['부실판단'] == 1]
bank2.drop_duplicates(subset='거래소코드', keep='first', inplace=True)

# 정상기업 데이터만 남기기
bank2_code = bank2['거래소코드'].unique() # 부실로 선정된 기업의 거래소코드
non_bank2 = df2[~df2['거래소코드'].isin(bank2_code)] # 부실선정 기업 제거
non_bank2.drop_duplicates(subset='거래소코드', keep='last', inplace=True)

# 부실기업과 정상기업 데이터프레임 합치기
final2 = pd.concat([bank2, non_bank2])
final2.reset_index(drop=True, inplace=True)
```

```python
final1 = final1.drop(columns=['Unnamed: 0', '감사의견코드', '[U01410000]영업활동으로 인한 현금흐름(*)(천원)','감사부실', '이보배부실', '현금흐름부실'])

final2 = final2.drop(columns=['Unnamed: 0', '감사의견코드', '[U01410000]영업활동으로 인한 현금흐름(*)(천원)', '감사부실', '이보배부실', '현금흐름부실'])
```
